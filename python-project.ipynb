# MGNREGA DATA ANALYSIS

## This dataset provides detailed insights into the implementation of the **Mahatma Gandhi National Rural Employment Guarantee Act (MGNREGA)** across various Indian states and districts from 2011 to 2022. It captures critical parameters such as:

- Number of job cards issued
- Households demanding and allotted work
- Person-days of employment generated (including gender-wise split)
- Financial disbursements (via banks and post offices)
- Participation of SC/ST communities
- District and state-level summaries

The dataset enables a deep dive into rural employment patterns, gaps in work allocation, and fund distribution efficiency — making it a powerful tool for both data-driven governance and academic exploration.

# 1. Identifying Districts with the Highest Employment Gaps under MGNREGA

```python
df.columns
```

```python
df.columns.to_list()
```

```python
df_gap = df[['srcstatename','srcdistrictname','srcyear','households_that_demanded_work','households_that_were_allotted_work']].copy()
df_gap.isnull().sum()
df_gap.dropna(inplace=True)
```

```python
df_gap['employment_gap'] = df_gap['households_that_demanded_work'] - df_gap['households_that_were_allotted_work']
df_gap.head()
```

```python
# Top 10 districts with highest employment gap
top_gap_districts = df_gap.sort_values(by='employment_gap', ascending=False)
top_gap_districts
top_gap_districts = df_gap.sort_values(by='employment_gap', ascending=False).head(10)
top_gap_districts
```

### Conclusion:

The analysis reveals that **all top 10 districts with the highest employment gaps under MGNREGA belong to the state of Andhra Pradesh**. This indicates a significant mismatch between the number of households demanding employment and those actually allotted work in the region. 

Such consistent under-allotment suggests possible systemic inefficiencies in MGNREGA implementation at the district level in Andhra Pradesh. Further investigation into administrative, logistical, or budgetary constraints in these districts could help bridge the employment gap and ensure more equitable job distribution under the scheme.

```python
# Group by State and calculate total employment gap
state_gap = df_gap.groupby('srcstatename')['employment_gap'].sum().reset_index()
state_gap = state_gap.sort_values(by='employment_gap', ascending=False)
# heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(state_gap.set_index('srcstatename'), annot=True, fmt=".0f", cmap='OrRd', linewidths=0.5, linecolor='gray')

plt.title('Total Employment Gap by State (MGNREGA)')
plt.ylabel('State')
plt.xlabel('Employment Gap')
plt.tight_layout()
plt.show()
```

### Heatmap of Employment Gaps by State

This heatmap shows the **total employment gap** across each state based on the difference between households that demanded work and those actually allotted work under MGNREGA.

🟧 **Darker colors** indicate larger employment gaps — visually confirming that **Andhra Pradesh** shows consistently high gaps compared to other states.

# 2.Investigate the participation of women in MGNREGA by analyzing total person-days worked by women across states and years.

```python
df_gender = df[['srcstatename', 'srcyear', 'total_person_days_worked_by_women']].copy()
df_gender.dropna(inplace=True)
total_by_state = df_gender.groupby(['srcstatename','srcyear'])['total_person_days_worked_by_women'].sum().reset_index()
total_by_state = total_by_state.sort_values(by=['srcstatename','srcyear'])
total_by_state
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))

for state in total_by_state['srcstatename'].unique():
    state_data = total_by_state[total_by_state['srcstatename'] == state]
    plt.plot(state_data['srcyear'], state_data['total_person_days_worked_by_women'], label=state)

plt.title('Women Participation in MGNREGA (Person-Days Worked)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Person-Days by Women in Millions')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Total participation by state (all years combined)
top_states = df_gender.groupby('srcstatename')['total_person_days_worked_by_women'].sum().reset_index()
top_states = top_states.sort_values(by='total_person_days_worked_by_women', ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(data=top_states.head(10), y='srcstatename', x='total_person_days_worked_by_women', palette='magma')
plt.title('Top 10 States by Women Participation in MGNREGA (All Years)', fontsize=13)
plt.xlabel('Total Person-Days Worked by Women in millions')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

### This section analyzes the **participation of women** in the MGNREGA scheme using total person-days worked across states and years.

- 📊 The **line plot** shows year-wise trends of women’s employment under MGNREGA across Indian states.
- 🥇 The **bar chart** ranks the top 10 states by their total women person-days worked — states like **Tamil Nadu, Andhra Pradesh, and Kerala** often top the list.

# 3: Efficiency of Fund Disbursement Mechanisms (Bank vs Post Office)

```python
cols = ['srcyear', 'srcstatename','amount_disbursed_to_bank_accounts','amount_disbursed_to_post_office_accounts','total_bank_accounts', 'post_office_accounts']
df_funds = df[['srcyear', 'srcstatename','amount_disbursed_to_bank_accounts','amount_disbursed_to_post_office_accounts','total_bank_accounts', 'post_office_accounts']].copy()
# Handle missing or none values
df_funds.fillna(0, inplace=True)
# Group by year and get totals
funds_summary = df_funds.groupby('srcyear')[['amount_disbursed_to_bank_accounts', 'amount_disbursed_to_post_office_accounts']].sum().reset_index()
funds_summary['amount_disbursed_to_bank_accounts'] /= 1_000_000
funds_summary['amount_disbursed_to_post_office_accounts'] /= 1_000_000
plt.figure(figsize=(12, 6))
plt.plot(funds_summary['srcyear'], funds_summary['amount_disbursed_to_bank_accounts'], marker='o', label='Bank Accounts')
plt.plot(funds_summary['srcyear'], funds_summary['amount_disbursed_to_post_office_accounts'], marker='s', label='Post Office Accounts')

plt.title('Fund Disbursement Trends via Bank vs Post Office (in ₹ Millions)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Amount Disbursed (in ₹ Millions)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

### CONCLUSION :
Bank accounts got way more funds than post office accounts every year. This shows people are shifting more towards digital banking, probably because it’s faster and easier.

# Objective 4: Forecasting Labour Demand Using Time-Series Trends

```python
labour_trend_df = df[['srcyear', 'households_that_demanded_work']].copy()
yearly_demand = labour_trend_df.groupby('srcyear')['households_that_demanded_work'].sum().reset_index()
yearly_demand.columns = ['Year', 'Total_Households_Demanded_Work']
yearly_demand['Total_Households_Demanded_Work'] /= 1_000_000
plt.figure(figsize=(10, 5))
plt.plot(yearly_demand['Year'], yearly_demand['Total_Households_Demanded_Work'], marker='o', color='teal')

plt.title('Labour Demand Trend Over the Years (in millions)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Households Demanded Work (in millions)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### CONCLUSION:
Over the years, the number of households demanding work under MGNREGA has steadily increased, highlighting a growing reliance on the scheme for rural employment support.

# Objective 5: Socio-Economic Impact
## Analyze how issuance of job cards (especially to SC/ST) correlates with actual work allocation and wage disbursement under MGNREGA.

```python

impact_df = df[['srcstatename','year','job_cards_issued_for_scheduled_caste','job_cards_issued_for_scheduled_tribes','households_that_were_allotted_work','labour_expenditure_that_has_been_disbursed']].copy()

# Drop missing values
impact_df.dropna(inplace=True)
impact_df.rename(columns={'job_cards_issued_for_scheduled_caste': 'SC_cards','job_cards_issued_for_scheduled_tribes': 'ST_cards','households_that_were_allotted_work': 'Work_allotted','labour_expenditure_that_has_been_disbursed': 'Labour_expenditure'}, inplace=True)

# Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(impact_df[['SC_cards', 'ST_cards', 'Work_allotted', 'Labour_expenditure']].corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("📊 Correlation: SC/ST Job Cards vs Work & Wage")
plt.tight_layout()
plt.show()

sc_trend = impact_df.groupby('year')[['SC_cards', 'Work_allotted']].sum().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(data=sc_trend, x='year', y='SC_cards', label='SC Job Cards Issued', marker='o')
sns.lineplot(data=sc_trend, x='year', y='Work_allotted', label='Work Allotted', marker='s')
plt.title("📈 SC Job Card Issuance vs Work Allotment Over Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### CONCLUSION
There's a visible positive correlation between job cards issued to SC/ST communities and work allotment. This suggests that MGNREGA is actively reaching marginalized groups. States with higher SC/ST outreach also tend to spend more on labour wages — a strong indicator of inclusive implementation.

# 6. Detecting Anomalies in Rural Employment Patterns Using Outlier Detection

```python
box_df = df[['srcstatename', 'srcdistrictname','households_that_demanded_work','households_that_were_allotted_work','labour_expenditure_that_has_been_disbursed']].dropna()

box_df.columns = ['State', 'District', 'Demanded', 'Allotted', 'Expenditure']
sns.set(style="whitegrid")

# Boxplot 1
plt.figure(figsize=(10, 4))
sns.boxplot(x=box_df['Demanded'], color='skyblue')
plt.title("Outlier Detection: Households That Demanded Work")
plt.xlabel("Households Demanded Work")
plt.tight_layout()
plt.show()

# Boxplot 2
plt.figure(figsize=(10, 4))
sns.boxplot(x=box_df['Allotted'], color='lightgreen')
plt.title("Outlier Detection: Households That Were Allotted Work")
plt.xlabel("Households Allotted Work")
plt.tight_layout()
plt.show()

#Boxplot 3
plt.figure(figsize=(10, 4))
sns.boxplot(x=box_df['Expenditure'], color='salmon')
plt.title("Outlier Detection: Labour Expenditure Disbursed")
plt.xlabel("Expenditure (in ₹)")
plt.tight_layout()
plt.show()
```
